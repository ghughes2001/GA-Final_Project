```python
# Importing the modules used for file handling
import pandas as pd
```

```python
# Function to read in data files
def file_reading(data):
    file = pd.read_csv(f"{data}")
    
    return file.head()
```

```python
coffee_data = file_reading("data/merged_data_cleaned.csv")
coffee_data.head()
```

```python
# looking at the statistics of the data set
coffee_data.describe()
```

```python
# Seeing if any information is missing
def missing_data(file):
    return file.isnull().head()

missing_data(coffee_data)
# some data is *NA*
```

```python
# Removing the missing data if not important when I analyze
coffee_data.drop("Lot.Number", axis = 1, inplace = True)
coffee_data.drop("Mill", axis = 1, inplace = True)
coffee_data.drop("ICO.Number", axis = 1, inplace = True)
coffee_data.drop("Color", axis = 1, inplace = True)
# Dropping some more columns that I am not going to use
coffee_data.drop("Region", axis = 1, inplace = True)
coffee_data.drop("Company", axis = 1, inplace = True) # This column also has a lot of NaN so bee better to remove it
coffee_data.drop("Producer", axis = 1, inplace = True) # Has a few N
coffee_data.drop("Farm.Name", axis = 1, inplace = True)
coffee_data.drop("Owner", axis = 1, inplace = True)
coffee_data.drop("Species", axis = 1, inplace = True) # Don't need species column since I am looking at origin
coffee_data.drop("Owner.1", axis = 1, inplace = True)
coffee_data.drop("Variety", axis = 1, inplace = True) # Not sure what this column is
coffee_data.drop("In.Country.Partner", axis = 1, inplace = True)
coffee_data.drop("Certification.Address", axis = 1, inplace = True)
coffee_data.drop("Certification.Contact", axis = 1, inplace = True)
coffee_data.drop("unit_of_measurement", axis = 1, inplace = True)
coffee_data.drop("Certification.Body", axis = 1, inplace = True)
coffee_data.drop("Processing.Method", axis = 1, inplace = True)
coffee_data.drop("Grading.Date", axis = 1, inplace = True)
coffee_data.drop("Altitude", axis = 1, inplace = True)
coffee_data.drop("Harvest.Year", axis = 1, inplace = True)
coffee_data.drop("Expiration", axis = 1, inplace = True)
```

```python
# Changes worked
```

```python
# Making the unmaded the index for dataframe since there is no need and does the same as default
coffee_data.set_index("Unnamed: 0", inplace = True)
```

```python
# Checking if changes are made
coffee_data.head()
```

```python
# Statistics for new data set
coffee_data.describe()
```

```python
# Plotting my data set to see what Country produces the most coffee in 2018
# *I want to see what the majority of data is coming from*
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12, 6))

plt.bar(coffee_data["Country.of.Origin"], coffee_data["Number.of.Bags"]);

# Add labels.
plt.title("Country that Produces the Most Coffee", size = 18)
plt.xlabel("Country", size = 10);
plt.ylabel("Sweetness", size = 10, rotation = 0, ha = "right");

plt.show()
```
