```python
# Imports needed
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to read in data files
def file_reading(data):
    file = pd.read_csv(f"{data}")
    
    return file.head()
```

```python
coffee_data = file_reading("data/Coffee_Data")
coffee_data
```

```python
# Need to fixe index column so going to make Unamed the index
coffee_data.set_index("Unnamed: 0", inplace = True)
```

```python
# Plotting my data set to see what Country produces the most coffee in 2018
# *I want to see what the majority of data is coming from*
fig2 = plt.figure(figsize = (12, 6))

plt.bar(coffee_data["Country.of.Origin"], coffee_data["Number.of.Bags"]);

# Add labels.
plt.title("Country that Produces the Most Coffee", size = 18)
plt.xlabel("Country", size = 10);
plt.ylabel("Sweetness", size = 10, rotation = 0, ha = "right");

plt.show()
```

```python
# Saving and exporting this file
fig2.savefig("img/Places_That_Produces_The_Coffee_In_DataFrame.png", facecolor = "#FFFFFF")
```

```python
# It seems like more places in Ethiopia produce a lot of quality coffee in 2018
# But I think that's because the data has mostly Ethiopia coffee.
```

```python
# Now going to plot according to "Aroma", "Flavor", "Aftertaste", "Acidity", "Body", "Balance", "Uniformity"
fig, ax = plt.subplots(4, 2, figsize = (20, 12), sharex = True);

ax[0][0].plot(coffee_data["Number.of.Bags"], coffee_data["Aroma"], linewidth = 3)
ax[0][0].set_title("Coffee Bags w/ Good Aroma", loc = "left", size = 20)

ax[0][1].plot(coffee_data["Number.of.Bags"], coffee_data["Flavor"], color = "red", linewidth = 3)
ax[0][1].set_title("Coffee Bags w/ Good Flavor", loc='left', size=20)

ax[1][0].plot(coffee_data["Number.of.Bags"], coffee_data["Aftertaste"], linewidth = 3)
ax[1][0].set_title("Coffee Bags w/ Good Aftertaste", loc = "left", size=20)

ax[1][1].plot(coffee_data["Number.of.Bags"], coffee_data["Acidity"], color = "purple", linewidth = 3)
ax[1][1].set_title("Coffee Bags And their Acidity", loc = "left", size = 20)

ax[2][0].plot(coffee_data["Number.of.Bags"], coffee_data["Body"], color = "purple", linewidth = 3)
ax[2][0].set_title("Coffee Bags And their Body", loc = "left", size = 20)

ax[2][1].plot(coffee_data["Number.of.Bags"], coffee_data["Acidity"], color = "purple", linewidth = 3)
ax[2][1].set_title("Coffee Bags And their Acidity", loc = "left", size = 20)

ax[3][0].plot(coffee_data["Number.of.Bags"], coffee_data["Balance"], color = "purple", linewidth = 3)
ax[3][0].set_title("Coffee Bags And their Balance", loc = "left", size = 20)

ax[3][1].plot(coffee_data["Number.of.Bags"], coffee_data["Uniformity"], color = "purple", linewidth = 3)
ax[3][1].set_title("Coffee Bags And their Uniformity", loc = "left", size = 20)

fig.tight_layout();

plt.show()
```

```python
# Saving the plots
# And setting background to be white, not transparent.
fig.savefig("img/Coffee_And_Their_Characteristics.png", facecolor = "#FFFFFF")
```
