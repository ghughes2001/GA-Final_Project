```python
# Imports needed
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to read in data files
def file_reading(data):
    file = pd.read_csv(f"{data}")
    
    return file
```

```python
coffee_data = file_reading("data/Coffee_Set")
coffee_data.head()
```

```python
# Need to fixe index column so going to make Unamed the index
coffee_data.set_index("Unnamed: 0", inplace = True)
```

```python
# Now going to plot according to "Aroma", "Flavor", "Aftertaste", "Acidity", "Body", "Balance", "Uniformity"
fig, ax = plt.subplots(4, 2, figsize = (20, 20));

ax[0][0].scatter(coffee_data["Number.of.Bags"], coffee_data["Aroma"], linewidth = 3)
ax[0][0].set_title("Coffee Bags w/ Good Aroma", loc = "left", size = 20)

ax[0][1].scatter(coffee_data["Number.of.Bags"], coffee_data["Flavor"], color = "red", linewidth = 3)
ax[0][1].set_title("Coffee Bags w/ Good Flavor", loc='left', size=20)

ax[1][0].scatter(coffee_data["Number.of.Bags"], coffee_data["Aftertaste"], linewidth = 3)
ax[1][0].set_title("Coffee Bags w/ Good Aftertaste", loc = "left", size=20)

ax[1][1].scatter(coffee_data["Number.of.Bags"], coffee_data["Acidity"], color = "purple", linewidth = 3)
ax[1][1].set_title("Coffee Bags And their Acidity", loc = "left", size = 20)

ax[2][0].scatter(coffee_data["Number.of.Bags"], coffee_data["Body"], color = "purple", linewidth = 3)
ax[2][0].set_title("Coffee Bags And their Body", loc = "left", size = 20)

ax[2][1].scatter(coffee_data["Number.of.Bags"], coffee_data["Acidity"], color = "purple", linewidth = 3)
ax[2][1].set_title("Coffee Bags And their Acidity", loc = "left", size = 20)

ax[3][0].scatter(coffee_data["Number.of.Bags"], coffee_data["Balance"], color = "purple", linewidth = 3)
ax[3][0].set_title("Coffee Bags And their Balance", loc = "left", size = 20)

ax[3][1].scatter(coffee_data["Number.of.Bags"], coffee_data["Uniformity"], color = "purple", linewidth = 3)
ax[3][1].set_title("Coffee Bags And their Uniformity", loc = "left", size = 20)

# fig.tight_layout();

plt.show()
```

```python
# It looks like the bags of coffee in this data is heavy in acidity, blanace, and body
```
