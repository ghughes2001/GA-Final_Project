```python
# importing the modules used for file handling
import pandas as pd
```

```python
# function to read in data files
def file_reading(data):
    file = pd.read_csv(f"{data}")
    
    return file.head()
```

```python
coffee_data = file_reading("data/merged_data_cleaned.csv")
coffee_data.head()
```

```python
# looking at the statistics of the data set
coffee_data.describe()
```

```python
# seeing if any information is missing
coffee_data.isnull().head()
```

```python
# removing the missing data if not important
coffee_data.drop("Lot.Number", axis = 1, inplace = True)
coffee_data.drop("Mill", axis = 1, inplace = True)
coffee_data.drop("ICO.Number", axis = 1, inplace = True)
coffee_data.drop("Color", axis = 1, inplace = True)
```
