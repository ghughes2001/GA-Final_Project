```python
# fixing error: *ValueError: You are trying to merge on object and int64 columns. If you wish to proceed you should use pd.concat*
set2["Harvest.Year"] = set2["Harvest.Year"].astype(str)
set2["Quakers"] = set2["Quakers"].astype(float)

# now time to merge the 2 sets together so I have one file for both
def file_merge(file1, file2):
    final_set = file1.join(file2)
    
    return final_set.head()

main_data_set = file_merge(set1, set2)
main_data_set
```
